### Joint Probabilites ( and | or) - Multiple Columns

### what is the probability of playing cricket and temparature is hot?

```python
sum((df['Play']=='Yes') & (df['Temperature']=='Hot'))/len(df)
```

```python
pd.crosstab(index=df['Play'],columns=df['Temperature'],normalize=True)
```

```python
pd.crosstab(index=df['Play'],columns=df['Temperature'],normalize=True).loc['Yes']['Hot']
```

### what is the probability of humidity is normal and temerature is cool?

```python
pd.crosstab(index=df['Humidity'],columns=df['Temperature'],normalize=True).loc['Normal']['Cool']
```

## Conditional Probability -- one is happend , what is the chance of other event 

## what is probability of playing cricket if it is raining

```python
pd.crosstab(index=df['Play'],columns=df['Outlook'],normalize=True)
```

```python
pd.crosstab(index=df['Play'],columns=df['Outlook'],normalize='columns').loc['Yes']['Rain']
```

## what is the probability of not playing cricket when humidity is high ?

```python
pd.crosstab(index=df['Humidity'],columns=df['Play'],normalize='index')
```

```python
pd.crosstab(index=df['Humidity'],columns=df['Play'],normalize='index').loc['High']
```

```python
pd.crosstab(index=df['Humidity'],columns=df['Play'],normalize='index').loc['High']['No']
```
